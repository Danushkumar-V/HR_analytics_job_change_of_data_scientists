```python
import pandas as pd
import numpy as np

train = pd.read_csv("C:/Users/danus/Downloads/hrdataset.csv")
```

```python
train
```

```python
train = train.drop(['enrollee_id', 'city', 'city_development_index'], axis = 1)
train.isnull().sum()
```

```python
train.dropna(subset = ['experience'], inplace = True, axis = 0)
train =train.fillna(method= "ffill")
train.dropna(subset = ['company_size', 'company_type'], axis = 0, inplace = True)
train.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
categorical = list(train.select_dtypes(include=["object"]).columns.values)
for cat in categorical:
    print(cat)
    train[cat].fillna('UNK', inplace=True)
    train[cat] = le.fit_transform(train[cat])
```

```python
for col in train.columns:
    train[col] = le.fit_transform(train[col])
```

```python
train.head(50)
```

```python
from sklearn.model_selection import train_test_split
x = train.drop('target', axis = 1)
y = train['target']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42) 
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import mean_absolute_error

knn = KNeighborsClassifier(n_neighbors =5)
knn.fit(x_train, y_train)    
y_pr = knn.predict(x_test)
# mean_absolute_error(y_test, y_pr)
knn_train_score = knn.score(x_train, y_train) 
knn_test_score = knn.score(x_test, y_test)
print(knn_train_score, knn_test_score)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100, random_state=0)
model.fit(x_train, y_train)
preds = model.predict(x_test)
mean_absolute_error(y_test, preds)

rf_train_score = model.score(x_train, y_train) 
rf_test_score = model.score(x_test, y_test)
print(rf_train_score, rf_test_score)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# Initiatlize the model
logreg = LogisticRegression(solver='liblinear', random_state = 0)

# Fit the model
logreg.fit(x_train, y_train)

# Predict data points 
y_pred_test = logreg.predict(x_test)

# mean_absolute_error(y_test,y_pred_test )

train_score = logreg.score(x_train, y_train)
test_score = logreg.score(x_test, y_test)
print(train_score, test_score)
```
